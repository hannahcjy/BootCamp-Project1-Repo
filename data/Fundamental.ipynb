```python
fundamental_df = pd.read_csv('ProjectOmega.csv')
sales2019 = fundamental_df['Net Sales']
sales2019.head(10)

```

```python
# If the data is in a dataframe, we use pandas to give quartile calculations
quartiles = sales2019.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of sales growth is: {lowerq}")
print(f"The upper quartile of sales growth is: {upperq}")
print(f"The interquartile range of sales growth is: {iqr}")
print(f"The median of sales growth is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
